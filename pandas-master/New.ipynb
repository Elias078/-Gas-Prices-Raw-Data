```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gasprices.csv')
gas.head(50)

```

```python
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'go-', label='Canada')
plt.plot(gas.Year, gas.UK, 'y--', label='UK')
#countries_to_look = ['Australia', 'USA', 'UK', 'Mexico']
#for country in countries_to_look:
#    if country != 'Year':
#        plt.plot(gas.Year, gas[country], marker='.')

plt.title('Gase prices 1990 To 2008')
plt.xticks(gas.Year[::2])
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()


plt.show()
```
